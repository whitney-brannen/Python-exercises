### What do all of the top colleges from the articles seem to have in common?
- High test scores, high median earning after 10 years, medium amount of debt

```python
# displaying common top colleges found in the articles
data = colleges[["INSTNM","STABBR","GRAD_DEBT_MDN","MD_EARN_WNE_P10","SAT_AVG","ADM_RATE","ACTCMMID"]]
data[data["INSTNM"].isin(["Massachusetts Institute of Technology","Stanford University","Princeton University","University of California, Berkeley","Columbia University","Harvard University","Yale University"])]
```

### First, I am creating a simple column I think is important for comparing the debt and loan amounts: "Income-Debt" (median income 10 years after graduation minus median debt after graduation)

```python
# Dropping PrivacySuppressed and Nan to convert entries to numeric values for subtraction calculation
data1 = data[data["GRAD_DEBT_MDN"]!= "PrivacySuppressed"].dropna()
data1 = data1[data1["MD_EARN_WNE_P10"]!= "PrivacySuppressed"].dropna()

# Creating debt to income column
debt_to_income = data1["MD_EARN_WNE_P10"].astype("float") - data1["GRAD_DEBT_MDN"].astype("float")

# Appending column to data
data1["Income-Debt"] = debt_to_income
data1.sort_values("Income-Debt",ascending=False)
```

### Next I am attempting to recreate the scoring system using 5 varables I believe are most important in this case:
    1. Median Earnings 10 years after graduation greater than $80,000
    2. Average SAT score of above 1490
    3. Median ACT score of 33 or higher
    4. An admission rate of less than 10%
    5. A high Income to debt range, where income is significantly higher than student loan debt

```python
# applying constraints
earnings = data1[data1["MD_EARN_WNE_P10"]>80000]
sat = earnings[earnings["SAT_AVG"]>1490]
act = sat[sat["ACTCMMID"]>=33]
adm = act[act["ADM_RATE"]<0.1]
overall = adm.sort_values(by="Income-Debt",ascending=False)
overall
```

## Determing cutoffs:
1. The average cost of tuition in the US is **19,020** a year. (https://www.bestcolleges.com/research/average-cost-of-college/)
- This leads me to look at schools `less than 30,000 a year`.
2. Public colleges have an average graduation rate of **63%**, and private nonprofit schools have a rate of **68%**. (https://www.bankrate.com/loans/student-loans/college-graduation-statistics/)
- This leads me to look at colleges with a degree completion rate of `at least 55%`.
3. The median annual salary for someone with a bachelors degree is **69,368**.  (https://www.bankrate.com/loans/student-loans/average-college-graduate-salary/)
- This leads me to look at schools where the median earnings after 10 years is `at least 60,000`.
4. The average college acceptance rate in the United States is **68%**. (https://research.com/education/list-of-college-acceptance-rates#:~:text=The%20average%20college%20acceptance%20rate,rate%20of%20around%20the%20average)
- This leads me to look at colleges with an acceptance rate of `at least 45%` (to have a realistic chance of getting in)
5. The 75th percentile of all SAT scores taken by highschool students is **1200**, which is typically considered acceptable for applying for most colleges. (https://www.bestcolleges.com/blog/what-is-a-good-sat-score/)
- I will be looking at schools with an average SAT score of `1100 or higher`. While colleges put a lot of weight into standardized test scores, it often does not reflect the actual classroom success of a student.

```python
# applying constraints
cost = part2[part2["COSTT4_A"]<=30000]
completion = cost[cost["C150_4"]>=.55]
salary = completion[completion["MD_EARN_WNE_P10"]>=65000]
acceptance = salary[salary["ADM_RATE"]>=.45]
sat = acceptance[acceptance["SAT_AVG"]>=1100]
overall2 = sat.sort_values(by="C150_4") # sorting by lowest cost of tuition
overall2
```
